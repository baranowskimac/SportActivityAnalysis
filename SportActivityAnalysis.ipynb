# Ocena jakości wizualizacji w raporcie "Sportkrastynacja 2024"

Choć wnioski przedstawione w raporcie są interesujące, jakość przynajmniej jednej z wizualizacji – tej na stronie 5, dotyczącej poziomu aktywności fizycznej Polaków – budzi istotne zastrzeżenia. Wykres prowadzi do ciekawych obserwacji, jednak jego forma graficzna utrudnia odbiór i może wprowadzać odbiorców w błąd. Poniżej przedstawiono główne problemy:

**Problemy z wizualizacją:**
1. Odwrócona chronologia:
Osie czasowe na wykresie przedstawiono w odwrotnej kolejności – najnowsze dane znajdują się po lewej stronie. Taki układ jest nieintuicyjny dla odbiorców przyzwyczajonych do czytania wykresów chronologicznych od lewej do prawej (czyli z danymi rosnącymi w czasie). Zmiana kolejności na: 2017 → ... → 2024 zwiększyłaby czytelność.
2. Błędy w skalowaniu i sumowaniu wartości:
Wykorzystano wykres słupkowy skumulowany (stacked bar chart), który zakłada, że suma kategorii dla każdego słupka wynosi 100%. W analizowanej wizualizacji wartości czasami przekraczają ten próg (np. 101%), a w innych przypadkach są niższe (np. 99%). Prawdopodobnie wynika to z błędów zaokrąglania, jednak dodatkowym problemem jest fakt, że słupki różnią się między sobą wysokością, co może sugerować różnice w liczebności próby – mimo że takiej informacji nie podano. To potencjalny błąd metodologiczny.
3. Nieprecyzyjna legenda:
Kategorie w legendzie są częściowo pokrywające się logicznie: np. „1–2 razy w tygodniu” oraz „1–3 razy w tygodniu”. Trudno ustalić, gdzie zakwalifikowano osoby ćwiczące 2 razy w tygodniu. Prawdopodobnie jedna z kategorii została błędnie opisana – istnieje też luka między „rzadziej niż raz w miesiącu” a „1–2 razy w tygodniu”, która powinna zostać uzupełniona (np. przez kategorię „1–2 razy w miesiącu”).
4. Nadmierna liczba kategorii:
Wykres zawiera zbyt wiele kategorii – co pogarsza jego czytelność. W przypadku wykresów typu stacked bar zaleca się ograniczenie liczby kategorii do maksymalnie 3–4, aby umożliwić szybkie porównania i ułatwić interpretację danych.

**Rekomendacje**

Wizualizacja powinna zostać poprawiona poprzez:

- zmianę kolejności lat na rosnącą (z lewej do prawej),
- korektę wartości procentowych tak, by każdy słupek sumował się dokładnie do 100%,
- doprecyzowanie kategorii w legendzie,
- ograniczenie liczby kategorii do kluczowych, łatwych do porównania.

Takie zmiany zwiększyłyby przejrzystość prezentowanych danych i pomogłyby uniknąć błędnych interpretacji w przyszłych raportach. 

# Propozycja poprawionej wizualizacji nr 2
Przy zachowaniu pełnej liczby kategorii i koncentracji na zmianach w czasie, rekomenduje się zastosowanie wykresu warstwowego (area chart). Tego typu wizualizacja umożliwia bardziej przejrzyste uchwycenie trendów, jednocześnie prezentując pełne spektrum odpowiedzi w sposób ciągły i czytelny.

**Przygotowanie danych i założenia:**
- Chronologiczne uporządkowanie – dane zostały uszeregowane od najstarszego (2017) do najnowszego (2024) roku, zgodnie z intuicyjnym kierunkiem odczytu (z lewej do prawej).
- Doprecyzowana legenda – poprawiono opisy kategorii, eliminując niejasności i potencjalne nakładanie się zakresów.
- Dodanie informacji o zmianach – przy końcu każdego pasma dodano adnotacje z informacją o zmianie wartości (w punktach procentowych) pomiędzy początkiem a końcem analizowanego okresu. Ułatwia to śledzenie dynamiki zmian.
- Usunięcie kategorii „Nie wiem” – ta kategoria pojawiła się wyłącznie w 2017 roku i dotyczyła 1% odpowiedzi. Ze względu na marginalny udział oraz wpływ na zaburzenie czytelności wykresu, zdecydowano się ją pominąć.
- Brak korekty sumowania do 100% – nie wprowadzano modyfikacji mających na celu pełne wyrównanie sum kategorii do 100%, ponieważ wykres oparto na danych przedstawionych w raporcie (bez dostępu do danych źródłowych). Bez znajomości dokładnych wartości bazowych nie było możliwe zaproponowanie rzetelnego rozwiązania metodologicznego

```python
# Dane przepisane z raportu multisport:
# wykres :Aktywni... coraz bardziej aktywni:, str. 5
# https://sportkrastynacja.pr1.cloud.benefitsystems.pl/pdf/Raport%20MS%20Index%202024%20Sportkrastynacja.pdf
ms_dict = {
    'Rok': ['2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024'],
    '5 razy w tygodniu lub częściej': [15, 16, 19, 20, 20, 19, 18, 20],
    '3 do 4 razy w tygodniu': [15, 14, 14, 16, 18, 18, 17, 16],
    '1 do 2 razy w tygodniu': [20, 24, 21, 21, 20, 20, 19, 21],
    '1 do 3 razy w miesiącu': [12, 8, 10, 8, 10, 8, 10, 9],
    'Rzadziej niż raz w miesiącu': [11, 9, 8, 8, 6, 8, 9, 6],
    'Nigdy': [26, 30, 29, 28, 27, 26, 26, 29],
    'Nie wiem': [1, 0, 0, 0, 0, 0, 0, 0]
}
ms_df = pd.DataFrame(ms_dict)
```

```python
# Jako, że kategoria "Nie wiem" pojawiła się tylko w jednym roku - należałoby ją usunąć z danych - brak porównań do innych lat
df_filtered = ms_df.drop(columns=["Nie wiem"])
df_filtered.set_index("Rok", inplace=True)
```

```python
#  Przygotowanie danych do wykresu
# Kolejność kolumn od najmniej aktywnych do najbardziej
column_order = [
    "Nigdy",
    "Rzadziej niż raz w miesiącu",
    "1 do 3 razy w miesiącu",
    "1 do 2 razy w tygodniu",
    "3 do 4 razy w tygodniu",
    "5 razy w tygodniu lub częściej"
]
df_filtered = df_filtered[column_order]

colors = [
    "#d8cfc4",  # Nigdy
    "#e3b7a0",  # Rzadziej niż raz w miesiącu
    "#f7c59f",  # 1–3 razy w miesiącu
    "#c7e8ca",  # 1–2 razy w tygodniu
    "#a7d7c5",  # 3–4 razy w tygodniu
    "#8fc1a9"   # 5 razy w tygodniu lub częściej
]
```

```python
# Styl seaborn
sns.set_style("whitegrid")

# przygotowanie wykresu
fig, ax = plt.subplots(figsize=(12, 7))
ax.stackplot(df_filtered.index, df_filtered.T.values, labels=column_order, colors=colors, alpha=0.8)

# Inline labels for category names in 2024
text_size = 10
position = 0.5
ax.text(
    x=position,
    y=15,
    s="Nigdy",
    ha='center',
    va='center',
    fontsize=text_size,
    color='black'
)

ax.text(
    x=position,
    y=33,
    s="Rzadziej niż raz w miesiącu",
    ha='center',
    va='center',
    fontsize=text_size,
    color='black'
)

ax.text(
    x=position,
    y=42,
    s="1 do 3 razy w miesiącu",
    ha='center',
    va='center',
    fontsize=text_size,
    color='black'
)

ax.text(
    x=position,
    y=58,
    s="1 do 2 razy w tygodniu",
    ha='center',
    va='center',
    fontsize=text_size,
    color='black'
)

ax.text(
    x=position,
    y=77,
    s="3 do 4 razy w tygodniu",
    ha='center',
    va='center',
    fontsize=text_size,
    color='black'
)

ax.text(
    x=position,
    y=92,
    s="5 razy w tygodniu lub częściej",
    ha='center',
    va='center',
    fontsize=text_size,
    color='black'
)


# Adnotacje różnic 2017 vs 2024
for i, category in enumerate(column_order):
    start = df_filtered.loc["2017", category]
    end = df_filtered.loc["2024", category]
    diff = end - start
    label = f"{diff:+.1f} p.p."
    cumulative = df_filtered.loc["2024"][:category].sum()-1
    # if diff < 0:
    ax.text(7.05, cumulative, label, va='center', ha='left', fontsize=12, color=colors[i])
    # else:
    #     ax.text(7.05, cumulative, label, va='center', ha='left', fontsize=12, color='#8fc1a9')

# Oś i legenda
ax.set_title("Zmiany w częstotliwości aktywności fizycznej Polaków (2017–2024)", fontsize=14)
ax.set_ylabel("")
ax.tick_params(axis='y', labelsize=14)
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_xlabel("")
ax.tick_params(axis='x', labelsize=14)
handles, labels = ax.get_legend_handles_labels()
ax.grid(axis='y', linestyle='--', alpha=0.9)
ax.grid(axis='x', linestyle='--', alpha=0.9)
plt.tight_layout()
plt.show()
```

# Propozycja poprawionej wizualizacji nr 2
W przypadku analizy liczebności poszczególnych grup, wybór wykresu typu stacked bar – zaproponowany w raporcie przez Benefit Systems S.A. – jest zasadny, jednak wymaga kilku istotnych poprawek w celu zwiększenia czytelności i poprawności interpretacyjnej.

**Wprowadzone zmiany:**
- Ograniczenie liczby kategorii do kluczowych grup:
    - Osoby uprawiające sport co najmniej raz w tygodniu zostały zakwalifikowane jako aktywni.
    - Osoby uprawiające sport 1–3 razy w miesiącu (poprawiona kategoria względem oryginalnej, błędnie opisanej jako "1–3 razy w tygodniu") sklasyfikowano jako sporadycznie aktywni – jako grupa pośrednia między nieaktywnymi a regularnie aktywnymi.
    - Kategorie „Nigdy” oraz „Rzadziej niż raz w miesiącu” połączono w jedną grupę nieaktywnych – z uwagi na to, że aktywność rzadsza niż raz w miesiącu ma charakter przypadkowy, a nie zaplanowany.
- Poprawa chronologii wykresu: Lata zostały uporządkowane w kolejności rosnącej (od 2017 do 2024), zgodnie z intuicyjnym kierunkiem odczytu danych.
- Usunięcie kategorii „Nie wiem”: Kategoria ta pojawiła się jedynie w 2017 roku i obejmowała 1% respondentów. Ze względu na jej marginalny udział oraz negatywny wpływ na czytelność wykresu, została pominięta.

```python
df_second = ms_df.copy()
df_second = pd.melt(df_second, id_vars='Rok', var_name='activity_type', value_name='perc_share')
df_second = df_second[df_second['activity_type'] != 'Nie wiem']
```

```python
cat = {
    '5 razy w tygodniu lub częściej': 'aktywni', 
    '3 do 4 razy w tygodniu': 'aktywni',
    '1 do 2 razy w tygodniu': 'aktywni', 
    '1 do 3 razy w miesiącu': 'sporadycznie aktywni',
    'Rzadziej niż raz w miesiącu': 'nieaktywni', 
    'Nigdy': 'nieaktywni'
}
df_second['cat'] = df_second['activity_type'].map(cat)
df_second.drop(['activity_type'], inplace=True, axis=1)
df_second = df_second.groupby(['Rok', 'cat']).sum('perc_share').reset_index().pivot(index=['Rok'], columns='cat', values = 'perc_share')[['nieaktywni', 'sporadycznie aktywni', 'aktywni']]
```

```python
# Pivot z odpowiednią kolejnością
# df_plot = df_grouped.pivot(index='Rok', columns='cat', values='perc_share')[['nieaktywni', 'sporadycznie', 'aktywni']]

# Styl Seaborn
sns.set_theme(style="whitegrid")

# Kolory pastelowe (spójne z narracją)
second_colors = ['#f4a582', '#fddbc7', '#8fc1a9']  # nieaktywni, sporadycznie, aktywni

# Tworzenie wykresu
fig, ax = plt.subplots(figsize=(12, 7))
bottom = [0] * len(df_second)
x = df_second.index

# Rysowanie każdej warstwy
for i, column in enumerate(df_second.columns):
    values = df_second[column].values
    ax.bar(x, values, bottom=bottom, label=column, color=second_colors[i], alpha=0.8)

    # Etykiety na środku każdej warstwy
    for j, val in enumerate(values):
        if val > 3:
            ax.text(j, bottom[j] + val / 2, f"{int(val)}%", ha='center', va='center', fontsize=12, color='#2e2e2e')

    # Uaktualnienie podstawy dla kolejnej warstwy
    bottom = [bottom[j] + values[j] for j in range(len(values))]

# Formatowanie
ax.set_title("Zmiany w częstotliwości aktywności fizycznej Polaków (2017–2024)", fontsize=16, pad=20)
ax.set_ylabel("")
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_xlabel("")
ax.tick_params(axis='x', labelsize=14)
ax.tick_params(axis='y', labelsize=12)
ax.legend(title="Poziom aktywności", loc="upper center", bbox_to_anchor=(0.5, -0.08), ncol=3, frameon=False, fontsize=12)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.grid(axis='x', visible=False)
plt.tight_layout()
plt.show()
```
